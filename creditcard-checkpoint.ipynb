## Handling the unscaled variables 
As two of the variables time and amount are not scaled, I scale them by using different libraries and make new variables 
to add in existing dataframe.

```python

from sklearn.preprocessing import StandardScaler, RobustScaler

std_scaler = StandardScaler()
rob_scaler = RobustScaler()

df['scaled_amount'] = rob_scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['scaled_time'] = rob_scaler.fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
```

```python

scaled_amount = df['scaled_amount']
scaled_time = df['scaled_time']     #### this will attach the scaled columns in dataframe

print(scaled_amount)
print (scaled_time)
df.drop(['scaled_amount', 'scaled_time'], axis=1, inplace=True)  #### this is drop and folowing inser command used to put 
                                                                ###scaled columns on 0 and 1 locations of table
df.insert(0, 'scaled_amount', scaled_amount)
df.insert(1, 'scaled_time', scaled_time)

# Amount and Time are Scaled!

df.head(100)
```

## Handling the Imbalanced dataset
There are many techniques such as under sampling, over sampling and SMOTE to handle the imbalanced data. If one performs
model fitting on imbalanced dataset then it will give overfitting or inf accuracy.

```python
df = df.sample(frac=1)

fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]   #### undersampling is done...data limits to 492 same as fraud

normal_distributed_df = pd.concat([fraud_df, non_fraud_df])  ##pd.concat adds the the colum with optional set along other axes

# Shuffle dataframe rows
new_df = normal_distributed_df.sample(frac=1, random_state=42)

new_df.head()
```

```python
print('Distribution of the Classes in the subsample dataset')
print(new_df['Class'].value_counts()/len(new_df))
```

```python
X = new_df.drop('Class', axis=1)
y = new_df['Class']
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df, y, test_size=0.2,random_state = 42)
X_train = X_train.drop('Class', axis=1)
X_test = X_test.drop('Class', axis=1)    ### class is dropped bcoz we are classifying it
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

## Logistic Regression (Detailed)

```python
logreg = LogisticRegression()
```

```python
logreg = logreg.fit(X_train,y_train)
training_score = cross_val_score(logreg,X_train,y_train,cv = 5)
print(training_score)
print("Logistic Regression has", round(training_score.mean(), 2) * 100, " % accuracy")
###y_pred=logreg.predict(X_train)
```

```python
y_pred1 = logreg.predict(X_test)
print(' Accuracy on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred1)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred1))
```

```python
training_scores = cross_val_score(logreg, X, y, cv=5, scoring='f1_macro')
training_scores

#### training score and f1-macro(score) showing the same accuracy because dataset is balanced. 
```

```python
from sklearn.metrics import roc_curve
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_auc_score
```

```python
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(false_positive_rate, true_positive_rate,label='Logistic Regression (area = %0.4f)' % logit_roc_auc)
plt.plot([0,1], [0,1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()
## blue color ROC Curve_ Curve of Probability (93% chance to classify between fraud and no_fraud)
### Higher the roc bettter the model i.e. better the classification
```

## Random Forest

```python
## Import the model we are using
from sklearn.ensemble import RandomForestClassifier

# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_train, y_train)

x_pred =rf.predict(X_train)
y_pred=rf.predict(X_test)
from sklearn import metrics
# Model Accuracy 
print("% train accuracy:",metrics.accuracy_score(y_train, x_pred))
print("% test accuracy:",metrics.accuracy_score(y_test, y_pred))
```
